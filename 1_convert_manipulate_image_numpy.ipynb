# Uploading and manipulating an image with numpy

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import PIL
```

```python
print('Installed Pillow Version:', PIL.__version__)
```

#### Get image details

```python
from PIL import Image

# import photo
image = Image.open('cat.jpg')
  
# summary image details
print(image.format)
print(image.size)
print(image.mode)
```

#### Convert JPEG to numpy array

```python
# convert to numpy array
from numpy import asarray

# load the image and convert into
# numpy array
img = Image.open('cat.jpg')
  
# Convert PIL image to numpy array using asarray()
numpydata = asarray(img)
  
# <class 'numpy.ndarray'>
print(type(numpydata))
  
#  shape
print(numpydata.shape)
```

```python
# view data
numpydata
```

```python
# view data as image
plt.imshow(numpydata)
```

#### Manipulate image with RGB image channels

```python
# the amount of red in the pixel at location (x, y)
print(numpydata[500, 500, 0])
# the amount of green in the pixel at location (x, y)
print(numpydata[500, 500, 1])
# the amount of blue in the pixel at location (x, y)
print(numpydata[500, 500, 2])
```

Use `vstack` to create a new image combined vertically

```python
plt.imshow(np.vstack([numpydata, numpydata]))
```

Or `hstack` to combine images horizontally:

```python
plt.imshow(np.hstack([numpydata, numpydata]))
```

Separate channels, then recombine using `dstack`:

```python
blank_image_channel = np.zeros((2148, 1975, 1), dtype=np.int64)

# create new images with one colour channel each
red_channel = numpydata[:, :, 0].reshape(2148, 1975, 1)
green_channel = numpydata[:, :, 1].reshape(2148, 1975, 1)
blue_channel = numpydata[:, :, 2].reshape(2148, 1975, 1)

# combine single channel images with blank image data to get images with single primary colours
red_image = np.dstack([red_channel, blank_image_channel, blank_image_channel])
green_image = np.dstack([blank_image_channel, green_channel, blank_image_channel])
blue_image = np.dstack([blank_image_channel, blank_image_channel, blue_channel])

plt.imshow(np.hstack([red_image, green_image, blue_image]))
```

Combine two channels:

```python
plt.imshow(np.dstack([blank_image_channel, green_channel, blue_channel]))
```

```python
plt.imshow(np.dstack([blank_image_channel, red_channel, blue_channel]))
```

```python
plt.imshow(np.dstack([blank_image_channel, green_channel, red_channel]))
```
